# __Ques 1__
Matplotlib is a Python library used for data visualization. It is a popular tool for creating a wide range of plots and charts, from basic line plots to complex 3D visualizations.
<br><br>
Matplotlib is used for several purposes such as:
1. Creating publication-quality figures for scientific publications or presentations.
2. Visualizing data to help understand trends, relationships, and patterns.
3. Exploring data interactively and iteratively.
4. Developing and testing machine learning models.
<br><br>
Here are five types of plots that can be created using the Pyplot module of Matplotlib:
1. __Line Plot:__ This is a basic plot type that displays data as a series of points connected by straight lines. It is useful for showing trends in data over time or across different categories.
2. __Scatter Plot:__ A scatter plot is used to visualize the relationship between two variables. It is a great tool for exploring correlations between data points.
3. __Bar Plot:__ A bar plot is used to compare data across different categories. It is useful for visualizing the distribution of a categorical variable.
4. __Histogram:__ A histogram is used to visualize the distribution of a numerical variable. It is useful for identifying patterns in data and detecting outliers.
5. __Pie Chart:__ A pie chart is used to visualize proportions of a whole. It is useful for showing how much of a given category makes up the total amount.

# __Ques 2__
A scatter plot is a type of plot in which the data is represented in the form of points where the valeu of points in x axis represenyt the value of x and the point in y axis represents the value of y axis

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter plot')
plt.show()
```

# __Ques 3__
The subplot() function is used to create multiple plots on a single figure. It takes three arguments: the number of rows, the number of columns, and the index of the subplot. The index starts from 1 and increments from left to right and top to bottom.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig , axis = plt.subplots(2,2)

axis[0,0].plot(x1,y1)
axis[0,0].set_title('Line 1')

axis[0,1].plot(x2,y2)
axis[0,1].set_title('Line 2')

axis[1,0].plot(x3,y3)
axis[1,0].set_title('Line 3')

axis[1,1].plot(x4,y4)
axis[1,1].set_title('Line 4')

fig.suptitle('Four line plots')

plt.show()
```

## __Ques 4__
A bar plot is a type of chart that displays rectangular bars with lengths proportional to the values they represent. Bar plots are used to compare categorical data or discrete values.
<br>
The x-axis of a bar plot represents the categories being compared, and the y-axis represents the values of the variable being measured. Each bar represents a category and is proportional in length to the value it represents.
<br>
Bar plots are useful for visualizing data that can be split into discrete categories, such as survey responses, test scores, or sales figures. They allow quick and easy comparison between categories and can reveal patterns or trends in the data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company , profit)
plt.title('Profit of each company')
plt.xlabel('Companies')
plt.ylabel('Profit')
plt.show()
```

A Box plot is a box-and-whiskers plot it have a box whose 2 ends whouse th2 first (first 25 %) and Third quartile(last 25 %) of the data, the linw between int the box shows the median of te data(first 50%) and the whiskers shows the maximum and minimum values which are not qutliers the data which grater than 1.5 times of third quartile and teh data which is 1.5 times less than the first quartile is known as outliers

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
fig , axis = plt.subplots(1,2)
axis[0].boxplot(box1)
axis[0].set_title('Box 1')

axis[1].boxplot(box2)
axis[1].set_title('Box 2')

fig.suptitle('Box-Whiskers-plot')
plt.show()
```
